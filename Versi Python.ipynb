```python
#Pertama kita import library yang akan dipakai
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#lalu kita masukkan dataset, kali ini saya meminjam dataset dari DQLab
url = "https://academy.dqlab.id/dataset/data_transaksi2.txt"
df = pd.read_csv(url, sep="\t")
```

```python
#Lihat data yang sudah di baca
df
```

```python
#Kita juga bisa melihat item-item unik, beserta jumlahnya, contoh dibawah item yang paling laku adalah shampo biasa yang telah terjual 604 kali, begitu juga barang yang paling jarang laku seperti Tali Pinggang anak yang hanya terjual sekali
df['Nama Barang'].value_counts()
```

```python
#Kita mulai data preprocessing, pertama saya encoding itemset agar berubah menjadi integer, karna algortima hanya membaca angka
transaksi = pd.get_dummies(df, columns=['Nama Barang'])
```

```python
#Lalu saya mengelompokkanya berdasarkan dengan kode transaksinya
transaksi = transaksi.groupby('Kode Transaksi').sum()
```

```python
#kita lihat data yang sudah kita proses sebelumnya
transaksi
```

```python
#Mari kita cek lebih mendetail, apakah ada baris yang kosong? Jika kita lihat keterangan di bawah, data sudah bagus dan siap untuk dipakai
transaksi.info()
```

```python
#Selanjutnya adalah membuat variabel dimana terdiri dari beberapa barang yang sering / terbeli dari seluruh transaksi menggunakan perintah apriori
itemset = apriori(transaksi, min_support=0.1, use_colnames=True)
```

```python
itemset
```

```python
#Lalu mulai melihat kecocokan assosiasi data, kali ini saya menggunakan parameter yang sama seperti yang saya gunakan di R, yaitu support confidence adalah 80%, karna di Python jika minimal kita set 80, maka angka 80 tidak akan dihitung, jadi saya ubah menjadi 79, agar support 80 terhitung
rules = association_rules(itemset, metric="confidence", min_threshold=0.79)
rules
```
